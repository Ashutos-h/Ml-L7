Now we need to perform Multivariate Analysis using the data from imported datsheet.
Multivariate analysis is performed by extracting more than 2 features from the datasheet at one time. 

```python
#Visualization 10----
sns.factorplot('Pclass', data=df, kind='count', hue='Sex', col='Survived', order=[1,2,3], 
               hue_order=['female','male'], aspect=1.25, size=5)
```

Inference drawn from above graph:
'''It is seen from the above graph that the number of females survived were more in each person class as the males tried to        rescue the females first and thus most of the rescue boats had females in them with less number of males able to escape.'''

Did Rose survive because she was a woman, or because she was in the upper class? Let’s have a look at passenger class. 
Here we see that 1st class passengers, or those who paid a higher fare, had greater chance of survival than the other classes.

```python
#Visualization 11----
sns.lmplot('Age', 'Survived', data=df)
```

'''There seems to be a general linear trend between age and the survived field. The plot shows that the older the passenger is,
   the less chance he/she would survive.'''

```python
#Visualization 12----
sns.lmplot('Age', 'Survived', data=df, hue='Sex')
```

'''Older women have higher rate of survival than older men as shown in the figure above. Also, older women has higher rate 
   of srvival than younger women; an opposite trend to the one for the male passengers.'''

```python
#Visualization 13----
sns.factorplot('Siblings/Spouses Aboard', col='Survived', data=df, kind='count', size=7, aspect=.8)
```

```python
#Visualization 14----
sns.factorplot('Parents/Children Aboard', col='Survived', data=df, kind='count', size=7, aspect=.8)
```

```python
Inference drawn from above graph:
'''The ones with families onboard were not as fortunate as those travelling alone.'''
```

```python
#Visualization 15----
sns.jointplot(x='Age', y='Fare',\
              data=df, kind='scatter')
```

```python
Inference drawn from above graph:
'''The highest fares were paid by the youth aged between 20-30 i.e. the person class contained mostly of youth'''
```

```python
#Intresting Graphs from the above analysis:
sns.lmplot('Age', 'Survived', data=df, hue='Sex')
sns.factorplot('Pclass', data=df, kind='count', hue='Sex', col='Survived', order=[1,2,3], 
               hue_order=['female','male'], aspect=1.25, size=5)
```
